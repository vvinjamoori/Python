```python
import time
import numpy as np

# Define a large size for our data structures
size = 10000000  # 10 million elements

# --- Python List ---

# Create a Python list
python_list = list(range(size))

# Measure time for element-wise multiplication
start_time = time.time()
python_list_result = [x * 2 for x in python_list]
end_time = time.time()
python_list_time = end_time - start_time
print(f"Time taken for Python list operation: {python_list_time:.4f} seconds")

# --- NumPy Array ---

# Create a NumPy array
numpy_array = np.arange(size)

# Measure time for element-wise multiplication
start_time = time.time()
numpy_array_result = numpy_array * 2
end_time = time.time()
numpy_array_time = end_time - start_time
print(f"Time taken for NumPy array operation: {numpy_array_time:.4f} seconds")

print(f"\nNumPy array was {python_list_time / numpy_array_time:.2f} times faster than Python list.")
```

### Why NumPy arrays are faster:

1.  **Memory Layout**: NumPy arrays are stored in a contiguous block of memory, which allows for highly efficient CPU cache utilization and vectorization (applying operations to entire arrays at once).
2.  **Fixed Type**: All elements in a NumPy array must be of the same data type. This homogeneity simplifies data storage and processing, as Python doesn't have to check the type of each element during an operation.
3.  **Optimized C Implementation**: The core operations in NumPy are implemented in C, which provides significant performance advantages over Python's interpreted code. This means that numerical computations are performed at a lower level, much closer to the hardware.
4.  **Vectorization**: NumPy encourages 'vectorized' operations, where operations are applied to the entire array rather than iterating through elements one by one (as is common with Python lists). This eliminates the overhead of Python loops, which can be very slow.

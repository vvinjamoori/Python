## **1. Introduction to Data Visualization with Matplotlib and Seaborn 🚀**

**📚 Context:**

Now, we will introduce two of the most popular Python libraries for data visualization: Matplotlib and Seaborn.

You will learn how these tools can help transform complex datasets, like our E-commerce Review Dataset, into insightful visual representations. Let's begin with a solid foundation!

### **1. What is Data Visualization? 📊**

Data visualization is the art of representing data graphically. It allows us to:  
- **Summarize** large datasets efficiently.  
- **Spot patterns, trends, and anomalies**.  
- **Communicate insights to others**.  
- **Support decision-making processes**.  

### Why Do We Need Data Visualization? 🧐  
1. **Simplify Complex Data:** Our dataset has 30 columns and 1737 rows. Without visuals, identifying relationships and distributions is challenging.  
2. **Highlight Insights:** For example, **understanding the distribution of ratings** or spotting **outliers in price** is much easier visually.  
3. **Support Hypotheses:** Visuals help validate our assumptions (e.g., Are higher-priced products better rated?).  

### **2. Key Libraries for Visualization**

### Matplotlib 📈  
**Matplotlib** is a versatile library for creating 2D plots and graphs in Python.  
- **Modules:** The `pyplot` module is like the "Swiss Army knife" of visualization—it provides tools for making basic plots like line graphs, bar charts, and scatter plots.  
- **Customizability:** Highly customizable but can be verbose for more advanced visuals.

### Seaborn 🎨  
**Seaborn** is built on top of Matplotlib and is tailored for statistical data visualization.  
- **Enhanced Aesthetics:** Seaborn’s visuals are clean, professional, and easy to interpret.  
- **Pandas Integration:** It works seamlessly with Pandas DataFrames, making it ideal for datasets like ours.  
- **Advanced Features:** Built-in support for advanced plots like heatmaps, pair plots, and categorical plots.  

💡 **Key Difference:** Matplotlib is great for full control and custom plots, while Seaborn simplifies and enhances plotting for statistical data.

> **Simple mental model for Figure and Axis:**
> Think of a plot like a **page**:
> * The full page is called the **figure**
> * Inside the page, one or more chart areas can be placed — each is called an **axis**
> * Inside each axis, we draw lines, bars, labels, and legends
> * A figure can contain: one chart, multiple charts, title, x-axis label, y-axis label, legend, and tick labels
>
> **Quick reference:**
> | Term | Meaning |
> | --- | --- |
> | Figure | The complete plotting area |
> | Axis | One chart area inside the figure |
> | X label | Label for the horizontal axis |
> | Y label | Label for the vertical axis |
> | Legend | Guide explaining chart elements |
> | Ticks | Values shown on the axis |

Woah! There is a lot of information in this image. Let's understand them one at a time.

- Figure: The **overall window** or page that everything is drawn on.
  - You can create multiple independent Figures in Jupyter.
  - If you run the code in terminal, separate windows will pop-up.
- Axes: You can add multiple **Axes** to the Figure, which represents a plot.

- **Axis**: Simply the `x-axis` and `y-axis`

- **Axes**: It is the **area** on which the **data is plotted** with functions such as `plot()`.
    - **x-label**: Name of x-axis
    - **y-label**: Name of y-axis

- **Major ticks**:
  - Subdivides the axis into major units.
  - They appear by default during plotting.
- **Minor ticks**:
  - Subdivides the major tick units.
  - They are by default hidden and can be toggled on.

- **Title**: Title of each plot **(Axes)**

- **Subtitle**: The common title of all the plots.

- **Legend**:
  - Describes the elements in the plot.
  - Blue and Green curves in this case.

These are the major components of a matplotlib plot.

# **Univariate Data Visualization**  

We start with univariate plots to understand **distributions** and **frequencies** of single columns.

### 1 📊 Visualization Goal: Histogram
A **histogram** helps visualize the distribution of a numerical column. Let’s explore the distribution of <font color="green">`rating`</font>.

#### Key Insights
- Identify skewness (are ratings generally high or low?).
- Check for outliers or unusual spikes in values.

### Code:

```python
!gdown 1WBGIqLNkmi_pWhC6i5tmfnTDARoPSaxd
```

```python
import pandas as pd
df = pd.read_csv('amazon_data.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Histogram for 'rating'
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='rating', kde=False, color='skyblue')

# Styling Tip 💡
# Add a vertical line for the mean rating
mean_rating = df['rating'].mean()
plt.axvline(mean_rating, color='red', linestyle='--', label=f"Mean: {mean_rating:.2f}")

plt.title("Distribution of Ratings", fontsize=16)
plt.xlabel("Rating", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.legend()
plt.show()
```

Equivalent version in Matplotlib:

### Code:

```python
plt.figure(figsize=(8, 6))
plt.hist(df['rating'].dropna(), bins=10, color='skyblue')
plt.axvline(mean_rating, color='red', linestyle='--', label=f"Mean: {mean_rating:.2f}")
plt.title("Distribution of Ratings")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

> **Seaborn vs Matplotlib:**
> * **Seaborn** generally produces cleaner, more visually appealing statistical plots with less code
> * **Matplotlib** gives more manual control over every plot element
> * Both are useful — Seaborn for quick statistical charts, Matplotlib when fine-grained customization is needed
> * Common plotting steps: (1) Create figure size → (2) Plot the chart → (3) Add title → (4) Add axis labels → (5) Add legend if needed → (6) Show

### <span style=color:orange> Business Context:</span>

The histogram of product ratings provides crucial insights into customer satisfaction. For an e-commerce platform like Amazon:

- A high concentration of 4–5 star ratings may indicate that products are meeting or exceeding customer expectations.

- A skew towards lower ratings could highlight quality issues, misleading product descriptions, or poor customer experience.

- Identifying outliers (e.g., unusually low ratings) helps the business investigate specific products or batches that may require attention.

- The mean rating line provides a quick benchmark to assess the overall product performance and compare across categories or brands.

A bar plot compares different categories, such as the count of products in each <font color="green">category</font>.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group by 'month_year' and calculate average rating
monthly_rating = df.groupby('month_year')['rating_filled'].mean().reset_index()

# Ensure 'month_year' is in datetime format
#sept - b
#September
monthly_rating['month_year'] = pd.to_datetime(monthly_rating['month_year'], format='%B %Y')

# Sort the DataFrame based on 'month_year.'
monthly_rating = monthly_rating.sort_values('month_year')

# Convert back to string for proper x-axis labels
monthly_rating['month_year'] = monthly_rating['month_year'].dt.strftime('%B %Y')

# Plot
plt.figure(figsize=(10, 6))
plt.plot(monthly_rating['month_year'], monthly_rating['rating_filled'], marker='o', color='navy')

# Rotate the x-axis labels and add a grid
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
#plt.grid(linestyle='--', alpha=0.7)

plt.title("Average Rating by Month-Year", fontsize=16)
plt.xlabel("Month-Year", fontsize=12)
plt.ylabel("Average Rating", fontsize=12)
plt.tight_layout()
plt.show()
```

**Instructor Note:** Notice the following line of code:

```python
monthly_rating['month_year'] = pd.to_datetime(monthly_rating['month_year'], format='%B %Y')
```

We convert `monthly_rating['month_year']` to a `datetime` type so that the data can be sorted in chronological (increasing month-wise) order. This ensures that the monthly ratings are analyzed in the correct sequence over time.

The format `'%B %Y'` is used so that the data is in the form of full month name followed by the year (e.g., *January 2023*), where:

* `%B` represents the full month name
* `%Y` represents the four-digit year



### Code:

```python
# Group by 'month_year' and calculate average rating
monthly_rating = df.groupby('month_year')['rating_filled'].mean().reset_index()

# Ensure 'month_year' is in datetime format
#sept - b
#September
monthly_rating['month_year'] = pd.to_datetime(monthly_rating['month_year'], format='%B %Y')

# Sort the DataFrame based on 'month_year.'
monthly_rating = monthly_rating.sort_values('month_year')

# Convert back to string for proper x-axis labels
monthly_rating['month_year'] = monthly_rating['month_year'].dt.strftime('%B %Y')

# Plot
plt.figure(figsize=(10, 6))
plt.plot(monthly_rating['month_year'], monthly_rating['rating_filled'], marker='o', color='navy')

# Rotate the x-axis labels and add a grid
plt.xticks(rotation=45)
plt.grid(False)
plt.tight_layout()#"Please adjust margins so everything fits neatly within the figure."
```

## <span style=color:skyblue>Visualization Goal: Grouped Bar Plot</span>
A grouped bar plot compares multiple categories side by side. Let’s compare the average `rating` for products that do or do not contain the keyword "durable", across different `rating_category`.

### Key Insights
* Check how “durable” items perform in each rating category vs. non-durable.
* Identify which group has higher/lower average ratings.

```python
# Create a pivoted dataframe
grouped_data = df.groupby(['rating_category', 'contains_durable'])['rating_filled'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=grouped_data, x='rating_category', y='rating_filled', hue='contains_durable', palette='Paired')

# Styling Tip 💡
# Add legend title
plt.legend(title='Contains Durable?')

plt.title("Average Rating by Rating Category and Durability", fontsize=16)
plt.xlabel("Rating Category", fontsize=12)
plt.ylabel("Average Rating", fontsize=12)
plt.show()
```

- Durable products generally receive slightly higher ratings—especially in the “Excellent” category—so emphasizing durability in product descriptions and marketing can help boost customer perception and ratings.
- The absence of durable products in the “Poor” category reinforces that low-quality, non-durable items are more likely to receive negative feedback, highlighting the importance of product longevity in customer satisfaction.

### <span style=color:orange> Business Context:</span>

A grouped bar plot comparing average ratings of durable vs. non-durable products across rating categories provides Amazon with insights into how product characteristics influence customer satisfaction:

- Durability impact: Products labeled as “durable” generally receive higher ratings, indicating that longevity and quality are key drivers of customer satisfaction.

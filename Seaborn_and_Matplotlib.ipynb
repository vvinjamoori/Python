```python
import pandas as pd
df = pd.read_csv('amazon_data.csv')
```

```python
df
```

```python
df['rating']
```

```python
#1-5 rating is spread across 1724 products
#how is that spread looking like

import seaborn as sns
import matplotlib.pyplot as plt
```

```python


plt.title("Distribution of Ratings", fontsize=12)

sns.histplot( data= df, x='rating')
plt.xlabel("Rating")
plt.ylabel("Frequency")

plt.show()
```

```python
sns.countplot(data=df, x='category') #this will give the wront because it is counting the duplicates as well
```

```python
cf= df.groupby('category')['product_id'].nunique().reset_index()
```

```python
cf
```

```python
sns.barplot(x="category", y="product_id", data=cf)
plt.xticks(rotation=90)
plt.ylabel("No. of Products")
plt.show()
```

```python
df.info()
```

```python
df['order_timestamp']= pd.to_datetime(df['order_timestamp'])
```

```python
df.info()
```

```python
df['month']= df['order_timestamp'].dt.month
```

```python
df
```

```python
cf= df.groupby('month')['rating'].mean().reset_index()
```

```python
cf #keep it in mind to sport the values if required
```

```python
sns.lineplot(x="month", y="rating",
             data=cf)
```

```python
#is there any trend or pattern between discounted price and ratings


sns.scatterplot(data=df, y='discounted_price', x='rating')
```

```python
# correlation between actual and discounted price
sns.scatterplot(data=df, y='discounted_price', x='actual_price')
```

```python
sns.scatterplot(data=df, y='discounted_price', x='actual_price', hue='category')
```

```python
cf= df.groupby(['category','contains_durable'])['product_id'].nunique().reset_index()
```

```python
cf
```

```python
sns.barplot(x="category", y="product_id",hue='contains_durable', data=cf)
plt.xticks(rotation=90)
plt.ylabel("No. of Products")
plt.show()
```

```python
def extract_price(price_str):
    try:
        return float(str(price_str).replace('₹', '').replace(',', ''))
    except ValueError:
        return None
```

```python
df['actual_price_fixed']= df['actual_price'].apply(extract_price)
```

```python
df['discounted_price_fixed']= df['discounted_price'].apply(extract_price)
```

```python
df
```

```python
# calculate the discount percentage for the product items
def calculate_discount_percentage(ef):
    return ((ef['actual_price_fixed'] - ef['discounted_price_fixed']) / ef['actual_price_fixed']) * 100

# Apply the function along rows (axis=1)
df['calculated_discount'] = df[['actual_price_fixed','discounted_price_fixed']].apply(calculate_discount_percentage, axis=1)
```

## Further Data Insights

### 1. Top 10 Products by Review Count

Let's identify the products that have the most reviews, which can be an indicator of their popularity or market interest.

```python
# Group by product name and sum the rating counts to find the most reviewed products
top_reviewed_products = df.groupby(['product_name', 'category'])['rating_count'].sum().nlargest(10).reset_index()
display(top_reviewed_products.style.background_gradient(cmap='Blues'))
```

### 2. Product with the Highest Discount

Understanding which products offer the highest discounts can reveal strategic pricing or clearance items.

```python
# Find the product with the maximum calculated discount
highest_discount_product = df.loc[df['calculated_discount'].idxmax()]
display(highest_discount_product[['product_name', 'category', 'actual_price_fixed', 'discounted_price_fixed', 'calculated_discount']])
```

```python
def categorize_rating(rating):
    if rating < 4:
        return 'low'
    else:
        return 'high'

categorize_rating_np= np.vectorize(categorize_rating) ############

ratings = np.array([4.9, 4.1, 4.8, 3.1,2.5,3.3])

categorize_rating_np(ratings)
```

```python
def discount_cost(rate, cost):
    "Apply a 10% discount if the rating is 4.0 or higher."""
    if rate >= 4.0:
        return cost * 0.9
    else:
        return cost

discount_cost_v= np.vectorize(discount_cost) #############
```
